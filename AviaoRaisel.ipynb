```
from random import Random
from time import time
from math import cos
from math import pi
from inspyred import ec
from inspyred.ec import terminators
import numpy as np
import os

def generate_(random, args):
    size = args.get('num_inputs', 12)
    return [random.randint(0, 15000) for i in range(size)]

def evaluate_(candidates, args):
    fitness = []
    for cs in candidates:
        fit = perform_fitness(cs[0], cs[1], cs[2], cs[3], cs[4], cs[5], cs[6], cs[7], cs[8], cs[9], cs[10], cs[11])
        fitness.append(fit)
    return fitness


def perform_fitness(C1D, C1C, C1T, C2D, C2C, C2T, C3D, C3C, C3T, C4D, C4C, C4T):

    C1D = np.round(C1D)
    C1C = np.round(C1C)
    C1T = np.round(C1T)

    C2D = np.round(C2D)
    C2C = np.round(C2C)
    C2T = np.round(C2T)

    C3D = np.round(C3D)
    C3C = np.round(C3C)
    C3T = np.round(C3T)

    C4D = np.round(C4D)
    C4C = np.round(C4C)
    C4T = np.round(C4T)
    
    C1 = C1D + C1C + C1T
    C2 = C2D + C2C + C2T
    C3 = C3D + C3C + C3T
    C4 = C4D + C4C + C4T

    CD = C1D + C2D + C3D + C4D
    CC = C1C + C2C + C3C + C4C
    CT = C1T + C2T + C3T + C4T

    n = 10;

    fit = float((0.31*C1 + 0.38*C2 + 0.35*C3 + 0.285*C4) / 12350)

    # restrinção peso por carga
    h1 = np.maximum(0, (C1-18000)) / float(18000/n)
    h2 = np.maximum(0, (C2-15000)) / float(15000/n)
    h3 = np.maximum(0, (C3-23000)) / float(23000/n)
    h4 = np.maximum(0, (C4-12000)) / float(12000/n)

    # restrinção volume por setor do avião
    h5 = np.maximum(0, float((0.48*C1D + 0.65*C2D + 0.58*C3D + 0.39*C4D) - 6800)) / float(6800/n)
    h6 = np.maximum(0, float((0.48*C1C + 0.65*C2C + 0.58*C3C + 0.39*C4C) - 8700)) / float(8700/n)
    h7 = np.maximum(0, float((0.48*C1T + 0.65*C2T + 0.58*C3T + 0.39*C4T) - 5300)) / float(5300/n)

     # restrinção proporção por compartimento
    h8 = np.maximum(0, float((CD/34000) - (10/34))) / float((10/34)/n)
    h9 = np.maximum(0, float((CC/34000) - (16/34))) / float((16/34)/n)
    h10 = np.maximum(0, float((CT/34000) - (8/34))) / float((8/34)/n)

    fit = fit - (h1 + h2 + h3 + h4 + h5 + h6 + h7 + h8 + h9 + h10)

    return fit

def solution_evaluation(C1D, C1C, C1T, C2D, C2C, C2T, C3D, C3C, C3T, C4D, C4C, C4T):

    C1D = np.round(C1D)
    C1C = np.round(C1C)
    C1T = np.round(C1T)

    C2D = np.round(C2D)
    C2C = np.round(C2C)
    C2T = np.round(C2T)

    C3D = np.round(C3D)
    C3C = np.round(C3C)
    C3T = np.round(C3T)

    C4D = np.round(C4D)
    C4C = np.round(C4C)
    C4T = np.round(C4T)

    C1 = C1D + C1C + C1T
    C2 = C2D + C2C + C2T
    C3 = C3D + C3C + C3T
    C4 = C4D + C4C + C4T

    CD = C1D + C2D + C3D + C4D
    CC = C1C + C2C + C3C + C4C
    CT = C1T + C2T + C3T + C4T

    print("-----RESUMO DA PRODUÇÃO-----")
    print("LUCRO TOTAL {}".format((0.31*C1 + 0.38*C2 + 0.35*C3 + 0.285*C4)))
    print(" ")

    print("-----TOTAL DE CARGAS UTILIZADAS:")
    print("Carga C1 (Max. 18000) {}".format(C1))
    print("Carga C2 (Max. 15000) {}".format(C2))
    print("Carga C3 (Max. 23000) {}".format(C3))
    print("Carga C4 (Max. 12000) {}".format(C4))
    
    print(" ")

    print("-----TOTAL DE CARGAS POR COMPARTIMENTO:")
    print("Carga compartimento Dianteiro (Max. 10000) {}".format(CD))
    print("Carga compartimento Central (Max. 16000) {}".format(CC))
    print("Carga compartimento Trazeiro (Max. 8000) {}".format(CT))
    print("Total (Max. 34000) {}".format(C1+C2+C3+C4))

    print(" ")

    print("-----PROPORÇÃO POR COMPARTIMENTO:")
    print("Carga compartimento Dianteiro (Max. 0.294117647) {}".format(CD/34000))
    print("Carga compartimento Central (Max. 0.470588235) {}".format(CC/34000))
    print("Carga compartimento Trazeiro (Max. 0.235294118) {}".format(CT/34000))

    print(" ")

    print("-----TOTAL VOLUME CARREGADO POR COMPARTIMENTO:")
    print("Volume compartimento Dianteiro (Max. 6800) {}".format(C1D*0.48 + C2D*0.65 + C3D*0.58 + C4D*0.39))
    print("Volume compartimento Central (Max. 8700) {}".format(C1C*0.48 + C2C*0.65 + C3C*0.58 + C4C*0.39))
    print("Volume compartimento Trazeiro (Max. 5300) {}".format(C1T*0.48 + C2T*0.65 + C3T*0.58 + C4T*0.39))
    print("Total (Max. 20800) {}".format(C1*0.48 + C2*0.65 + C3*0.58 + C4*0.39))

    print(" ")

    print("-----DISTRIBUIÇÃO DE CARGAS POR COMPARTIMENTO:")
    print("Carga C1 - Dianteira: {} - Central: {} - Traseira: {}".format(C1D, C1C, C1T))
    print("Carga C2 - Dianteira: {} - Central: {} - Traseira: {}".format(C2D, C2C, C2T))
    print("Carga C3 - Dianteira: {} - Central: {} - Traseira: {}".format(C3D, C3C, C3T))
    print("Carga C4 - Dianteira: {} - Central: {} - Traseira: {}".format(C4D, C4C, C4T))


def main():
    rand = Random()
    rand.seed(int(time()))

    ea = ec.GA(rand)
    ea.selector = ec.selectors.tournament_selection
    ea.variator = [ec.variators.uniform_crossover,
                   ec.variators.gaussian_mutation]
    ea.replacer = ec.replacers.steady_state_replacement

    ea.terminator = terminators.generation_termination
    ea.observer = [ec.observers.stats_observer, ec.observers.file_observer]

    final_pop = ea.evolve(generator=generate_,
                          evaluator=evaluate_,
                          pop_size=1000,
                          maximize=True,
                          bounder=ec.Bounder(0, 23000),
                          max_generations=15000,
                          max_inputs=12,
                          crossover_rate=1.0,
                          num_crossover_points=1,
                          mutation_rate=0.2,
                          num_elites=1,
                          num_selected=12,
                          tournament_size=12,
                          statistics_file=open('aviaoRaisel.csv', 'w'),
                          individuals_file=open('aviaoRaisel_individuals.csv', 'w'))

    final_pop.sort(reverse=True)
    print(final_pop[0])

    perform_fitness(final_pop[0].candidate[0], final_pop[0].candidate[1], final_pop[0].candidate[2], final_pop[0].candidate[3],
                    final_pop[0].candidate[4], final_pop[0].candidate[5], final_pop[0].candidate[6], final_pop[0].candidate[7],
                    final_pop[0].candidate[8], final_pop[0].candidate[9], final_pop[0].candidate[10], final_pop[0].candidate[11])
    
    solution_evaluation(final_pop[0].candidate[0], final_pop[0].candidate[1], final_pop[0].candidate[2], final_pop[0].candidate[3],
                    final_pop[0].candidate[4], final_pop[0].candidate[5], final_pop[0].candidate[6], final_pop[0].candidate[7],
                    final_pop[0].candidate[8], final_pop[0].candidate[9], final_pop[0].candidate[10], final_pop[0].candidate[11])

main()
```
